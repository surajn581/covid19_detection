```python
from shutil import copyfile
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
src = 'D:/Downloads/gta5/Compressed/covid_19_dataset/'
dst = 'D:/Downloads/gta5/Compressed/covid_19_dataset_sorted/data/'
```

```python
data_details = pd.read_csv(src+'labels.csv')
```

```python
data_details.head()
```

```python
print(data_details.View.unique())
print(data_details.label.unique())
```

```python
img = cv2.imread(src+data_details.image[0])
plt.imshow(img)
```

```python
for i in range(2248,len(data_details)):
    if data_details.View[i] == 'Front':
        if data_details.label[i] == 'non-infection':
            try:
                copyfile(src+data_details.image[i], dst+'healthy/'+data_details.image[i].replace('/','-'))
            except:
                print(i, ' File error')
        elif data_details.label[i] == 'covid-19 pneumonia':
            try:
                copyfile(src+data_details.image[i], dst+'covid/'+data_details.image[i].replace('/','-'))
            except:
                print(i, ' File error')
        elif data_details.label[i] == 'non-covid-19 pneumonia':
            try:
                copyfile(src+data_details.image[i], dst+'noncovid/'+data_details.image[i].replace('/','-'))
            except:
                print(i, ' File error')
        else:
            print('No conditions matched')
    else:
        print('View is not front')
    print(i, ' is done')
print('Copy Complete')
```
